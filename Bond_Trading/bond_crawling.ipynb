```python
dk = ecals.get_calendar('XKRX', start='20010404', end='20230521', side='left')
```

```python
dt.datetime.strftime(dk.schedule.index[0], '%Y%m%d')
```

```python
from tqdm import tqdm
```

```python
df_list = []
idx_list = []

for x in tqdm(dk.schedule.index[dk.schedule.index > '2002-01-01']):
    idx = dt.datetime.strftime(x, '%Y%m%d')
    df_list.append(bond.get_otc_treasury_yields(idx)[['수익률']])
    idx_list.append(x)

df = pd.concat(df_list, keys=idx_list).unstack()
df.columns = df.columns.get_level_values(1)
df
```
